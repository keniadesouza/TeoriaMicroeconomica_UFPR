# REVISÃO: Máximos e mínimos

## Máximo

Como exemplo para máximo vamos usar: 

$F(x) = -x^2 + 40x$

Plotando a função: 

```python
from sympy import *

x = symbols('x')
F = -x**2 + 40*x
plot(F, (x, 0, 50), ylim = (0, 500), xlim = (0, 55), xlabel = "x", ylabel = "F(x)")
```

Primeira derivada: 

```python
diff(F, x)
```

Encontrando o ponto de máximo: 

$F'(x) = 0$

```python
xmax = solve(Eq(diff(F, x),0))
print(" x = ", xmax[0])
```

```python
Fxmax = F.subs(x, xmax[0])
print("F(20) = ", Fxmax)
```

```python
plot(F, Fxmax, (x, 0, 50), ylim = (0, 500), xlim = (0, 55), xlabel = "x", ylabel = "F(x)")
```

Para um ponto de máximo local, é condição suficiente que a segunda derivada seja menor do que zero: 

$F''(x) < 0$

Segunda derivada: 

```python
diff(diff(F, x))
```

Logo, trata-se de um **máximo local**. 

## Mínimo

Como exemplo para mínimo, vamos usar: 

$F(x) = 0.05x^2 - 2x + 50$

```python
%reset -f
from sympy import *
x = symbols('x')
F = 0.05*x**2 - 2*x + 50
plot(F, (x, -50, 100), ylim = (0, 200), xlim = (-50, 100), axis_center = (0, 0), xlabel = "x", ylabel = "F(x)")
```

Primeira derivada: 

```python
diff(F, x)
```

Condição de primeira ordem ($F'(x)=0$): 

```python
xmax = solve(Eq(diff(F, x),0))
```

Solução da condição de primeira ordem: 

```python
print(" x = ", xmax[0])
Fxmax = F.subs(x, xmax[0])
print("F(20) = ", Fxmax)
```

Condição de segunda ordem: 

```python
diff(diff(F, x))
```

Como $F''(x)>0$, trata-se de um mínimo local.

# Multiplicador de Lagrange

Exemplo: 

Um consumidor tem preferências descritas pela função $𝑈(𝑥_1,𝑥_2)=𝑥_1^{0.5}𝑥_2^{0.5}$, sendo os preços dos bens $𝑥_1$ e $𝑥_2$    representados por $𝑝_1$ e $𝑝_2$ e a renda por $𝑚$. Utilizando o método de lagrange, encontre a demanda ótima. 

Ou seja, nesse caso:

$𝑈(𝑥_1,𝑥_2)=𝑥_1^{0.5}𝑥_2^{0.5}$

$p_1x_1 + p_2x_2 = m$

Declarando as variáveis: 

```python
x1, x2, p1, p2, m, l = symbols('x1, x2, p1, p2, m, l', positive = True)
```

Função de utilidade:

```python
U = x1**0.5*x2**0.5
```

Lagrangeano:

```python
L = U - l*(p1*x1 + p2*x2 - m)
```

Condições de primeira ordem: 

```python
E1 = Eq(diff(L,x1), 0) 
E2 = Eq(diff(L,x2), 0)
E3 = Eq(diff(L, l), 0)
```

Resultado: 

```python
solve((E1, E2, E3), [x1, x2, l])
```
