# Teoria Microeconômica - Monopólio

## Exemplo numérico utilizado em aula

Função de demanda:

$ P = 40 - Q $

Custo total do monopolista: 

$ CT = 50 + Q^2$

```python
%reset -f
from sympy import *
Q = symbols('Q')
P = 40 - Q
RT = P*Q
CT = 50 + Q**2
RMg = diff(RT, Q)
CMg = diff(CT, Q)
Qr = solve(Eq(RMg, CMg))
Qr
```

Outra forma de resolver:

```python
Lucro = RT - CT
Qr = solve(diff(Lucro, Q))
Qr
```

```python
CMe = CT/Q
plot(RMg, P, CMg, CMe, (Q, 0, 50), ylim = (0, 50), xlim = (0, 50), xlabel = "Quantidade", ylabel = "Preço", legend=True)
```

Lucro total: 

```python
Lucro.subs(Q, Qr[0])
```

## Determinação dos preços em um monopólio

A regra de determinação dos preços, derivada em sala de aula, é dada por: 

$$
P = \frac{CMg}{1-\frac{1}{E_d}}
$$

Considerando $CMg = 10$, o que ocorre com o preço quando a elasticidade da demanda varia?

**Observação: nos gráficos e fórmulas abaixo (para facilitar a interpretação) vamos considerar o módulo da elasticidade-preço da demanda.**

```python
%reset -f
from sympy import *
Ed, CMg, MKup = symbols('Ed, CMg, MKup')
CMg = 10
P = CMg/(1-(1/Ed))
```

```python
Pp = plot(P, (Ed, 0, 10),
          ylim = (-100, 100), xlim = (0, 10),
          axis_center = (0, 0), 
          xlabel = "Módulo da Elasticidade-Preço", ylabel = "Preço", 
          legend = True, 
          size = (9,5),
          show = False
        )
Pp[0].label = 'Preço'
CMgp = plot(CMg, (Ed, 0, 10), 
          ylim = (-100, 100), xlim = (0, 10),
          axis_center = (0, 0), 
          xlabel = "Módulo da Elasticidade-Preço", ylabel = "Preço", 
          legend = True, 
          show = False
        )
CMgp[0].label = 'CMg'
Pp.append(CMgp[0])
Pp.show()
```

Observe que quando $|E_d|>1$ (demanda é elástica em relação aos preços), temos $P>CMg$. Nesse trecho, $P$ tende ao infinito quando $|E_d|$ se aproxima de $1$ e tende a se igualar ao custo marginal quando a elasticidade-preço aumenta. 

Por outro lado, quando $|E_d|=1$, o preço seria infinito, e quando $|E_d|<1$ (demanda inelástica), o preço seria menor do que zero. 

Obviamente, o monopolística irá operar apenas em trechos de demanda elástica.
